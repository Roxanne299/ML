```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1,2,3,4,5,6,7,8,9,0]
plt.plot(y,lw=4,c='r',ls='--',aa=True)   #aa参数就是让曲线变得平滑
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
print(lines)
line = lines[0]
line.set_color('r')
line.set_linewidth(4)
line.set_alpha(0.5)
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
plt.setp(lines,linewidth=4,color = "r")
```

# 笔记：
1. 使用plot绘图的时候，就可以传递Line2D的属性值进去，来修改线条的样式。
2. 使用plot返回的线条，单独去设置也可以，但是单独设置的时候，是通过`set_Line2D属性名()`的方式来设置的。
3. 使用plt.setp来一次性设置多个线条的样式。
