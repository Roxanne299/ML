```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
values = np.arange(20)
plt.subplot(221)
plt.plot(values)
plt.plot(values**2)
plt.subplot(222)
plt.plot(np.sin(values),'r')
plt.subplot(223)
plt.plot(np.cos(values),'b')
plt.subplot(224)
plt.plot(np.tan(values),'y')
```

```python
values = np.linspace(0,20)
plt.style.use("seaborn")
fig,axes = plt.subplots(2,2,sharex=True) #Axes
print(axes)
ax1 = axes[0,0]
ax1.plot(values)
ax2 = axes[0,1]
ax2.plot(np.sin(values),c='y')
ax3 = axes[1,0]
ax3.plot(np.cos(values),c='b')
ax4 = axes[1,1]
ax4.plot(np.tan(values),c='g')
```
