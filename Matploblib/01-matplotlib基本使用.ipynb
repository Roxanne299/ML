```python
mydata = {
    "a": range(5,25),
    "b": [np.random.randint(0,10) for x in range(20)]
}
mydf = pd.DataFrame(data=mydata)
# print(mydf)
plt.plot("a","b","",data=mydf) # range(0,10)
```
